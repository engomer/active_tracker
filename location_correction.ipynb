```python
import requests
import json
username = "bc143a"
server_url = "http://activetrackerv1.herokuapp.com/api/getLocation.php?user=" + username
corrector_url = "http://router.project-osrm.org/route/v1/driving/"
correct_url_config = "?steps=true&geometries=geojson"

server_answer = requests.get(server_url)
parsed = json.loads(server_answer.content)
```

```python
import datetime

curTime = parsed[0]["time"].split(":")
curTime = datetime.time(int(curTime[0]),int(curTime[1]),int(curTime[2]))
curDate = parsed[0]["date"].split("/")
curDate = datetime.date(int(curDate[0]),int(curDate[1]),int(curDate[2]))
arr = []
for i in range(1,len(parsed)):
    nowTime = parsed[i]["time"].split(":")
    nowTime = datetime.time(int(nowTime[0]),int(nowTime[1]),int(nowTime[2]))
    nowDate = parsed[i]["date"].split("/")
    nowDate = datetime.date(int(nowDate[0]),int(nowDate[1]),int(nowDate[2]))
    if(curDate.year == nowDate.year and curDate.month == nowDate.month and curDate.day == nowDate.day):
        if(nowTime.minute > curTime.minute + 1 or (nowTime.hour != curTime.hour and nowTime.minute  > (curTime.minute + 61))):
            print(str(nowDate) + "," + str(nowTime))
            print(str(curDate) + "," + str(curTime))
            print("\n")
            #OSRM çalışacak
            nowLat = parsed[i]["lattitude"]
            nowLong = parsed[i]["longtitude"]
            coord = str(curLong) + "," + str(curLat) + ";" + str(nowLong) + "," + str(nowLat)
            correction_path = requests.get(corrector_url + coord + correct_url_config)
            parsed_path = json.loads(correction_path.content)
            #print(parsed_path["routes"][0]["geometry"]["coordinates"])
            routed_coord = parsed_path["routes"][0]["geometry"]["coordinates"]
            for j in range(len(routed_coord)):
                arr.append([float(routed_coord[j][1]) , float(routed_coord[j][0])])
            
        else:
            arr.append([float(parsed[i]["lattitude"]), float(parsed[i]["longtitude"])])
            
    if(curDate != nowDate and nowTime.hour != 0):
        print(str(nowDate) + "," + str(nowTime))
        print(str(curDate) + "," + str(curTime))
        print("\n")
        nowLat = parsed[i]["lattitude"]
        nowLong = parsed[i]["longtitude"]
        coord = str(curLong) + "," + str(curLat) + ";" + str(nowLong) + "," + str(nowLat)
        correction_path = requests.get(corrector_url + coord + correct_url_config)
        parsed_path = json.loads(correction_path.content)
        #print(parsed_path["routes"][0]["geometry"]["coordinates"])
        routed_coord = parsed_path["routes"][0]["geometry"]["coordinates"]
        for j in range(len(routed_coord)):
            arr.append([float(routed_coord[j][1]) , float(routed_coord[j][0])])
        #osrm çalışcak
    else:
        arr.append([float(parsed[i]["lattitude"]), float(parsed[i]["longtitude"])])

    curTime = parsed[i]["time"].split(":")
    curTime = datetime.time(int(curTime[0]),int(curTime[1]),int(curTime[2]))
    curDate = parsed[i]["date"].split("/")
    curDate = datetime.date(int(curDate[0]),int(curDate[1]),int(curDate[2]))
    curLat = parsed[i]["lattitude"]
    curLong = parsed[i]["longtitude"]
```
